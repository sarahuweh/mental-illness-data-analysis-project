<font color='green'>**DATASET**</font>

This is an exploratory data analysis project. The data is an open source data gotten from [Kaggle](https://www.kaggle.com/datasets/osmi/mental-health-in-tech-survey). The single csv file containing the data was downloaded from the site. The data revels the results from a mental health survey carried out on tech companies employees. The data usability score on Kaggle was 8.82. The data is licensed by [CC BY-SA 4.0](https://creativecommons.org/licenses/by-sa/4.0/). The data is anonymous as it does not in any way lead to the tracing of the individual respondents.


The original data was collected by [Open Sourcing Mental Illness (OSMI)](https://osmhhelp.org/index.html) which is a non-profit corporation dedicated to raising awareness, educating, and providing resources to support mental wellness in the tech and open source communities.
OSMI began in 2013 as a speaking campaign by Ed Finkler who spoke about his personal experiences as a web developer at tech conferences. Ed is also an open source advocate with a mental health disorder. The response was overwhelming, and Ed speaks, gather data, and organize efforts to change experiences of those with mental health disorders in the tech workplace.

I checked out a dataset on student's mental health on [Kaggle](https://www.kaggle.com/datasets/shariful07/student-mental-health). It was a more recent survey as it was carried out in 2020 but when I looked through the data, I discovered it did not have any checks on the attitude of respondents in relation to treatment of mental illness. I also checked this data set on Global Trends in Mental Health Disorder on [Kaggle](https://www.kaggle.com/datasets/thedevastator/uncover-global-trends-in-mental-health-disorder), discovered there were no parameters used to measure if the respondent have ever sought treatment on mental health illness.

The data used on Mental Health in Tech Survey [Kaggle](https://www.kaggle.com/datasets/osmi/mental-health-in-tech-survey) has 27 columns and 1259 rows. Python was my preferred programming language as Python allows mandatory and optional arguments, keyword arguments, and even arbitrary argument lists. I imported libraries and made use of the following libraries; (*pandas, seaborn, matplotlib, numpy*), to enable quick and effiecient data manipulation and analysis. I also viewed the heading of the data to take a glance at the column headings so I could understand the data collected and available for analysis.




**Introduction to Dataset**

Below is an explanation of the parameters used for the survey.

*Column Description*

- **Timestamp**        - The date and time in which the respondent took the survey
- **Age**              - Age of the respondent
- **Gender**	         - Gender of the respondent
- **Country**	         - The country of residence of the respondent
- **state**	         - The state or territory the respondent lives in
- **self_employed**    - Is the respondent self employed
- **family_history**   - Does the respondent have a family history of mental illness
- **treatment**        - Has the respondent sought treatment for a mental health condition
- **work_interfere**   - Has a mental health condition interfered with the respondent's work
- **no_employees**     - The number of employees in the respondent's company
- **remote_work**      - Does the respondent work remotely
- **tech_company**     - Does the respondent work in a tech company
- **benefits**         - Does the respondent's company provide mental health benefits
- **care_options**     - Does the respondent know about their company's mental health care benefits
- **wellness_program** - Has the respondent's employer discussed mental health as part of an employee wellness program
- **seek_help**        - Has the respondent's employer provided resources for employees to educate themselves about mental health and seek help
- **anonymity**        - Will the respondent remain anonymous if they decide to seek help for mental health
- **leave**            - Is it easy for the respondent to take medical leave for a mental health issue
- **mental_health_consequence** - Would dicussing a mental health issue with an employer result in a negative consequence
- **phys_health_consequence**   - Would dicussing a physical health issue with an employer result in a negative consequence
- **coworkers**                 - Is the respondent willing to discuss a mental health issue with their coworkers
- **supervisor**                - Is the respondent willing to discuss a mental health issue with their direct supervisor
- **mental_health_interview**   - Would the respondent bring up a mental health issue with a potential employer during an interview
- **phys_health_interview**     - Would the respondent bring up a physical health issue with a potential employer during an interview
- **mental_vs_physical**        - Does the respondent's employer take mental health as seriously as physical health
- **obs_consequence**           - Has the respondent observed any negative consequences for coworkers with mental health conditions
- **comments**                  - Additional comments

```python
#Creating Work Environment - Loading Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing CSV and inspecting data
Mental_health_df = pd.read_csv("survey.csv", encoding = "ISO-8859-1")
Mental_health_df.head()
```

```python
#Viewing the data shape ie number of rows and column
Mental_health_df.shape
```

<font color='green'>**DATA CLEANING**</font>


After viewing the headings of the data I decided to take out irrelevant data, check for inconsistencies, errors as well as missing values in the data.

- **Columns:** I removed irrelevant columns which will not be needed for my analysis. The *timestamp* had the date and time the data was collected hence not needed. I do not intend to draw any insights from the *states*, *self_employed* while the *comments* were optional and had different answers which would be difficult to analyse, so I dropped them. I realised that *work_interfere* column had lots of missing values and so in order to maintain the data integrity I decided to drop the column too.

- **Age:** When I viewed the unique ages, I discovered the had outliers which would have affected the analysis so I dropped them.I had to check again to confirm they were dropped.

- **Gender:** Upon viewing the unique genders in the dataset, I discovered we has lots of options, some were even wrongly spelt. This was due to the fact that the respondent were not given boxes to pick options from but were allowed to write out their response hence multiple responses with numerous errors. 

```python
# Dropping Irrelevant Columns
Mental_health_df.drop(columns=['Timestamp', 'state', 'comments', 'work_interfere', 'self_employed'], inplace = True)
```

```python
#Viewing the unique ages
Mental_health_df['Age'].unique()
```

```python
#Dropping outliers in age
Mental_health_df.drop(Mental_health_df[Mental_health_df['Age'] < 18].index, inplace = True)
Mental_health_df.drop(Mental_health_df[Mental_health_df['Age'] > 100].index, inplace = True);

#Confirming dropped outliers
Mental_health_df['Age'].unique()
```

```python
#Viewing unique gender in dataset
print(Mental_health_df['Gender'].unique())
```

```python
#Renaming gender options in dataset
Mental_health_df['Gender'].replace(['Male ', 'male', 'M', 'm', 'Male', 'Cis Male',
                     'Man', 'cis male', 'Mail', 'Male-ish', 'Male (CIS)',
                      'Cis Man', 'msle', 'Malr', 'Mal', 'maile', 'Make',], 'Male', inplace = True)

Mental_health_df['Gender'].replace(['Female ', 'female', 'F', 'f', 'Woman', 'Female',
                     'femail', 'Cis Female', 'cis-female/femme', 'Femake', 'Female (cis)',
                     'woman',], 'Female', inplace = True)

Mental_health_df["Gender"].replace(['Female (trans)', 'queer/she/they', 'non-binary',
                     'fluid', 'queer', 'Androgyne', 'Trans-female', 'male leaning androgynous',
                      'Agender', 'A little about you', 'Nah', 'All',
                      'ostensibly male, unsure what that really means',
                      'Genderqueer', 'Enby', 'p', 'Neuter', 'something kinda male?',
                      'Guy (-ish) ^_^', 'Trans woman',], 'Other', inplace = True)
```

```python
#Confirming changes made to gender
print(Mental_health_df['Gender'].unique())
```

```python
#Confirming that there is no missing value in the cleaned data
if Mental_health_df.isnull().sum().sum() == 0:
    print('There is no missing data')
else:
    print('There is {} missing data'.format(Mental_health_df.isnull().sum().sum()))
```

```python
#Viewing the summary information of the clean data to be used
print(Mental_health_df.info())
```

<font color='green'>**Treatment of Respondents**</font>

I decided to plot barcharts to compare the data of those who had sought mental health illness treatment and those who did to draw insights. I compared treatment against *family history*, *anonymity*, *obs_consequences*, *care_options* and *coworkers*.

```python
#Viewing respondents who have sought treatment for a mental health condition
sns.countplot(data = Mental_health_df , x = 'treatment', palette='Blues' )
```

```python
#Ploting Family History of Survey Respondents 
plt.figure(figsize = (20,6))
plt.subplot(1,2,1)
eda_percentage = Mental_health_df['family_history'].value_counts(normalize = True).rename_axis('family_history').reset_index(name = 'Percentage')
ax = sns.barplot(x = 'family_history', y = 'Percentage', data = eda_percentage, palette='Blues')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center', fontweight='bold')

plt.title('Family History of Survey Respondents', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

#Ploting Family History of Survey Respondents vs Treatment
plt.subplot(1,2,2)
sns.countplot(x=Mental_health_df['family_history'], hue = Mental_health_df['treatment'], palette='Blues')
plt.title('Family History of Survey Respondents vs Treatment', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

```python
#Ploting anonymity of survey respondents
plt.figure(figsize = (20,6))
plt.subplot(1,2,1)
eda_percentage = Mental_health_df['anonymity'].value_counts(normalize = True).rename_axis('anonymity').reset_index(name = 'Percentage')
ax = sns.barplot(x = 'anonymity', y = 'Percentage', data = eda_percentage, palette='Blues')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center', fontweight='bold')

#Ploting anonymity of survey respondents vs treatment
plt.title('Anonymity of Survey Respondents', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(1,2,2)
sns.countplot(x=Mental_health_df['anonymity'], hue = Mental_health_df['treatment'], palette='Blues')
plt.title('Anonymity of Survey Respondents vs Treatment', fontsize=18, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

```python
#Ploting obs_consequence of survey respondents
plt.figure(figsize = (20,6))
plt.subplot(1,2,1)
eda_percentage = Mental_health_df['obs_consequence'].value_counts(normalize = True).rename_axis('obs_consequence').reset_index(name = 'Percentage')
ax = sns.barplot(x = 'obs_consequence', y = 'Percentage', data = eda_percentage, palette='Blues')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center', fontweight='bold')

#Ploting obs_consequence of survey respondents vs treatment
plt.title('obs_consequence of Survey Respondents', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(1,2,2)
sns.countplot(x=Mental_health_df['obs_consequence'], hue = Mental_health_df['treatment'], palette='Blues')
plt.title('obs_consequence of Survey Respondents vs Treatment', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

```python
#Ploting care_options of survey respondents
plt.figure(figsize = (20,6))
plt.subplot(1,2,1)
eda_percentage = Mental_health_df['care_options'].value_counts(normalize = True).rename_axis('care_options').reset_index(name = 'Percentage')
ax = sns.barplot(x = 'care_options', y = 'Percentage', data = eda_percentage, palette='Blues')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center', fontweight='bold')

#Ploting care_options of survey respondents vs treatment
plt.title('care_options of Survey Respondents', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(1,2,2)
sns.countplot(x=Mental_health_df['care_options'], hue = Mental_health_df['treatment'], palette='Blues')
plt.title('care_options of Survey Respondents vs Treatment', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

```python
#Ploting coworkers of survey respondents
plt.figure(figsize = (20,6))
plt.subplot(1,2,1)
eda_percentage = Mental_health_df['coworkers'].value_counts(normalize = True).rename_axis('coworkers').reset_index(name = 'Percentage')
ax = sns.barplot(x = 'coworkers', y = 'Percentage', data = eda_percentage, palette='Blues')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center', fontweight='bold')

#Ploting coworkers of survey respondents vs treatment
plt.title('coworkers of Survey Respondents', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(1,2,2)
sns.countplot(x=Mental_health_df['coworkers'], hue = Mental_health_df['treatment'], palette='Blues')
plt.title('coworkers of Survey Respondents vs Treatment', fontsize=16, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

<font color='red'>**Insight on Respondents seeking Treatment**</font>

- **Family history:** It is seen that most people who don't have a family history of mental health issues had not sought treatment for a mental health condition while those who have sought treatment for mental health issues have a family history of mental illness.


- **Anonymity:** Most people who have sought treatment for mental health issues will remain anonymous if they decide to seek help for mental health.


- **obs_consequence:** People who have not sought treatment for mental health issues have not observed any negative consequences for coworkers with mental health conditions while those who have sought treatment have observed negative consequences.

- **Care_options:** People who have sought treatment for mental health issues tend to know about their company's mental health care benefits than those who haven't sought treatment.

- **coworkers:** There is really no difference with the willing to discuss a mental health issue with their coworkers among those who have sought mental health illness treatment and those who haven't
